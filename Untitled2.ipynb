# Dealing with our Categorical Variables

# Choosing a K value by using elbow method to pick a good k value

```python
error_rate=[]
for i in range(1,11):
    clf= KNeighborsClassifier(n_neighbors=i)
    clf.fit(Xtrain,Ytrain)
    Y_pred=clf.predict(Xtest)
    error_rate.append(np.mean(Y_pred!=Ytest))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error _rate vs k vaue')
plt.xlabel('k')
plt.ylabel('Error_rate')
```
